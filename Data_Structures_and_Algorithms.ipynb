### Linked List

[Linked lists](https://realpython.com/linked-lists-python/#implementing-your-own-linked-list) are an ordered collection of objects (nodes). The key part about them, is that each node is unaware of the nodes in the rest of the list, apart from its immediately adjacent neighbours. This makes removing/inserting nodes very straightforward, provided traversal to get to the node of a given point doesn't take too long (access is sequential by nature).

```python
class Node:
  def __init__(self,data):
    self.data = data
    self.next = None

  def __repr__(self):
    return self.data

class LinkedList:
  def __init__(self):
    self.head = None

  def __repr__(self):
    node = self.head
    nodes = []
    while node is not None:
      nodes.append(node.data)
      node = node.next
    nodes.append("None")
    return " -> ".join(nodes)

  def __iter__(self):
    node = self.head
    while node is not None:
      yield node
      node = node.next

  def push(self,node): # insert at start
    node.next = self.head
    self.head = node

  def append(self,node): # insert at end
    if self.head is None:
        self.head = node
        return
    for currentNode in self:
        pass
    currentNode.next = node

  def delete(self,targetNodeData):
    if self.head is None:
      raise Exception("List is empty")

    if self.head.data == targetNodeData:
      self.head = self.head.next
      return

    previous_node = self.head
    for node in self:
      if node.data == targetNodeData:
        previous_node.next = node.next
        return
      previous_node = node

    raise Exception("Node with data '%s' not found" % targetNodeData)

  # TODO: reverse a linked list using a stack
```

### Doubly Linked List

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
    self.prev = None

class DoublyLinkedList:
  def __init__(self):
    self.head = None
    self.tail = None

  def printList(self):
    currentNode = self.head
    while currentNode is not None:
      print(currentNode.data)
      currentNode = currentNode.next

  def __iter__(self):
    node = self.head
    while node is not None:
      yield node
      node = node.next

  def append(self, data): # add to end
    newNode = Node(data)
    if self.head is None:
      self.head = newNode
      self.tail = newNode
    else:
      newNode.prev = self.tail
      self.tail.next = newNode
      self.tail = newNode

  def delete(self, data): # consider next and prev now
    currentNode = self.head
    while currentNode is not None:
      if currentNode.data == data:
        if currentNode.data == self.head:
          self.head = currentNode.next
        else:
          currentNode.prev.next = currentNode.next
        if currentNode == self.tail:
          self.tail = currentNode.prev
        else:
          currentNode.next.prev = currentNode.prev
        currentNode = None
      elif currentNode.data < data:
        currentNode = currentNode.next
      else:
        break

dllist = DoublyLinkedList()
dllist.append(1)
dllist.append(5)
dllist.append(3)
dllist.printList()
dllist.delete(5)
dllist.printList()
```

### Hash Table

```python
# could use a DICT for this but will do class defintion too
# (more flexible for advanced problems, maybe)

class HashTable:
  def __init__(self, size):
    self.size = size
    self.table = [None] * self.size

  def hash(self, key):
    hashVal = 0
    for c in key:
      hashVal += ord(c) # get unicode int of char
    return hashVal % self.size

  def get(self, key):
    index = self.hash(key)
    node = self.table[index]

    while node is not None:
      if node.key == key:
        return node.value
      node = node.next

    return None

  def put(self, key, value):
    index = self.hash(key)
    node = self.table[index]

    while node is not None:
      if node.key == key:
        node.value = value
        return
      node = node.next

    newNode = Node(key, value)
    newNode.next = self.table[index]
    self.table[index] = newNode

class Node:
  def __init__(self, key, value):
    self.key = key
    self.value = value
    self.next = None

hashBrown = HashTable(10)
hashBrown.put("eggs", 1)
hashBrown.put("bacon", 2)

print(hashBrown.get("eggs"), hashBrown.get("bacon"))
```

### Binary Search Tree

```python
class BST:
  def __init__(self, data = None):
    self.left = None
    self.right = None
    self.data = data

  def insert(self, data):
    if self.data is None:
      self.data = data
      return

    if self.data == data:
      return

    if self.data > data:
      if self.left:
        self.left.insert(data)
        return
      self.left = BST(data)
      return

    if self.right:
      self.right.insert(data)
      return
    self.right = BST(data)

  def getMin(self):
    current = self
    while current.left is not None:
      current = current.left
    return current.data

  def getMax(self):
    current = self
    while current.right is not None:
      current = current.right
    return current.data

  def delete(self, data):
    if self is None:
      return self

    if data < self.data:
      self.left = self.left.delete(data)
      return self

    if data > self.data:
      self.right = self.right.delete(data)
      return self

    if self.right == None:
      return self.left

    if self.left == None:
      return self.right

    minLargerNode = self.right
    while minLargerNode.left:
      minLargerNode = minLargerNode.left
    self.val = minLargerNode.val
    self.right = self.right.delete(minLargerNode.val)

    return self

  def exist(self, data):
    if data == self.data:
      return True

    if data < self.data:
      if self.left == None:
        return False
      return self.left.exists(data)

    if self.right == None:
      return False

    return self.right.exists(data)

  def inorder(self, values):
    if self.left is not None:
      self.left.inorder(values)
    if self.data is not None:
      values.append(self.data)
    if self.right is not None:
      self.right.inorder(values)
    return values

  def preorder(self, values):
    if self.data is not None:
      values.append(self.data)
    if self.left is not None:
      self.left.preorder(values)
    if self.right is not None:
      self.right.preorder(values)
    return values

  def postorder(self, values):
    if self.left is not None:
      self.left.postorder(values)
    if self.right is not None:
      self.right.postorder(values)
    if self.data is not None:
      values.append(self.data)
    return values

# test
nums = [2, 4, 24, 1, 9, 20, 9, 5, 6, 8, 19]
bst = BST()
for num in nums:
  bst.insert(num)

print("preorder:")
print(bst.preorder([]))
print("#")

print("postorder:")
print(bst.postorder([]))
print("#")

print("inorder:")
print(bst.inorder([]))
print("#")
```

### Trie

https://albertauyeung.github.io/2020/06/15/python-trie.html/

CleanShot 2023-07-06 at 10.55.04@2x.png

```python
class Node:
  def __init__(self, char):
    self.char = char
    self.isEnd = False
    self.counter = 0
    self.children = {}

class Trie(object):
  def __init__(self):
    self.root = Node("")

  def insert(self, word):
    node = self.root

    # for each character
    for char in word:
      # check if no child contains the character
      if char in node.children:
        # if yes, make it a child for current node
        node = node.children[char]
      # else create new node for it in trie
      else:
        newNode = Node(char)
        node.children[char] = newNode
        node = newNode

    # end of word
    node.isEnd = True
    node.counter += 1

  def depthFirstSearch(self, node, prefix):
    if node.isEnd:
      self.output.append((prefix + node.char, node.counter))

    for child in node.children.values():
      self.depthFirstSearch(child, prefix + node.char) # recursion

  def query(self, x):
    # Given a prefix input, retrieve all words stored in the trie with it
    # Sort the words by the number of times they have been inserted
    self.output = []
    node = self.root

    # check prefix exists
    for char in x:
      if char in node.children:
        node = node.children[char]
      else:
        return []

    # traversal to get all candidates
    self.depthFirstSearch(node, x[:-1])

    # sort results in reverse order
    return sorted(self.output, key = lambda x: x[1], reverse=True)

t = Trie()
t.insert("had")
t.insert("have")
t.insert("heat")
t.insert("horse")
t.insert("house")
t.insert("happy")
t.insert("hustle")
t.query("ha")
```

### Quick Sort

https://www.delftstack.com/tutorial/algorithm/quick-sort/

```python
def quickSort(arr):
  left  = []
  right = []
  equal = []

  if len(arr) > 1:
    pivot = arr[0]
    for x in arr:
      if x < pivot:
        left.append(x)
      elif x == pivot:
        equal.append(x)
      elif x > pivot:
        right.append(x)

    # recursion
    return quickSort(left) + equal + quickSort(right)

  else: return arr

arr = [4,5,7,1,2,5,9,0,14,69,0,6,1,6,4]
print(quickSort(arr))
```

### Merge Sort

https://www.geeksforgeeks.org/merge-sort/

```python
def mergeSort(arr):
  if len(arr)>1:
    mid = len(arr)//2
    l = arr[:mid]
    r = arr[mid:]

    # recursion
    mergeSort(l); mergeSort(r)

    i = j = k = 0

    # copy to temp arrays l[] and r[]
    while i < len(l) and j < len(r):
      if l[i] <= r[j]:
        arr[k] = l[i]
        i += 1
      else:
        arr[k] = r[k]
        j += 1
      k += 1

    # check for remaining elements
    while i < len(l):
      arr[k] = l[i]
      i += 1
      k += 1
    while j < len(r):
      arr[k] = r[j]
      j += 1
      k += 1

    return(arr)

arr = [12, 11, 13, 5, 6, 7]
print("unsorted list:", arr)
print("after merge sort:", mergeSort(arr))
```

### Heap Sort

```python
import heapq
def heapSort(nums):
  heap = []
  for val in nums:
    # push elements to heap
    heapq.heappush(heap, val)
  # pop smallest elements and append to sorted heap
  return [heapq.heappop(heap) for i in range(len(heap))]

sortedList = heapSort([1,3,5,7,9,2,4,6,8,0])
print(sortedList)
```

### Insertion Sort

```python
def insertionSort(arr):
  for i in range(1,len(arr)):
    key = arr[i]
    t = i-1
    while t >= 0 and key < arr[t]:
      arr[t+1] = arr[t]
      t -= 1
    arr[t+1] = key

arr = [23,45,22,6,11]
insertionSort(arr)
for i in range(len(arr)):
  print("% d" % arr[i])
```

### Radix Sort

https://www.programiz.com/dsa/radix-sort

https://www.geeksforgeeks.org/radix-sort/

```python
# sorting digit-by-digit

def countingSort(arr, exp1):
  n = len(arr)
  outputArr = [0] * (n)
  countArr = [0] * (10)

  for i in range(0,n):
    idx = arr[i] // exp1
    countArr[idx % 10] += 1

  for i in range(1,10):
    countArr[i] += countArr[i - 1]

  i = n - 1
  while i >= 0:
    idx = arr[i] // exp1
    outputArr[countArr[idx % 10] - 1] = arr[i]
    countArr[idx % 10] -= 1
    i -= 1

  i = 0
  for i in range(0,len(arr)):
    arr[i] = outputArr[i]

def radixSort(arr):
  # get max value to know order of magnitude
  max1 = max(arr)

  # pass exp (10^i) where i = current digit number
  exp = 1
  while max1 / exp >= 1:
    countingSort(arr,exp)
    exp *= 10

arr = [170, 45, 75, 90, 802, 24, 2, 66]
radixSort(arr)
for i in range(len(arr)):
    print(arr[i],end=" ")
```

## Search

### Linear Search

The simplest search algorithm. Also one of the most inefficient, particularly as the dimension of the array increases.

```python
# iterative
def linSearch(arr,x):
  for i in range(len(arr)):
    if arr[i] == x:
      return i
  return -1

# recursive - versatile for higher dimensions
def linSearch2(arr,idx,x):
  if idx == -1:
    return -1
  if arr[idx] == x:
    return idx
  return linSearch2(arr,idx-1,x)

arr = [10, 20, 80, 30, 60, 50, 110, 100, 130, 170]
x = 110

print(linSearch(arr,x))

print(linSearch2(arr,len(arr)-1,x))
```

### Binary Search

```python
def binSearch(arr, low, high, x):
  if high >= low:
    mid = (high+low) // 2

    if arr[mid] == x:
      return mid

    # recursion
    elif arr[mid] > x:
      return binSearch(arr,low,mid-1,x)
    else:
      return binSearch(arr,mid+1,high,x)
  else:
    return -1

arr = [2,3,4,10,40]
x = 10
idx = binSearch(arr,0,len(arr)-1,x)

if idx != -1:
  print("element found at index", str(idx))
else:
  print("element not present in array")
```

## Traverse

### Minimum Spanning Tree

The graph that connects all nodes, with the lowest overall weight. Here are two [greedy algorithms](https://www.geeksforgeeks.org/introduction-to-greedy-algorithm-data-structures-and-algorithm-tutorials/).

#### Kruskal's Algorithm

https://www.geeksforgeeks.org/kruskals-minimum-spanning-tree-algorithm-greedy-algo-2/

```python
class Graph:
  def __init__(self, nodes):
    self.nodes = nodes
    self.graph = []

  def addEdge(self, u, v, w):
    self.graph.append([u, v, w])

  def find(self, parent, i):
    if parent[i] != i:
      parent[i] = self.find(parent, parent[i])
    return parent[i]

  def union(self, parent, rank, x, y):
    if rank[x] < rank[y]:
      parent[x] = y
    elif rank[x] > rank[y]:
      parent[y] = x
    else:
      parent[y] = x
      rank[x] += 1

  def kruskal(self):
    result = []
    i = 0 # sorted edges
    e = 0 # result edges

    # sort edges in ascending weight order
    self.graph = sorted(self.graph, key = lambda item: item[2])

    parent, rank = [], []

    # create subsets with single elements
    for node in range(self.nodes):
      parent.append(node)
      rank.append(0)

    while e < self.nodes - 1:

      u, v, w = self.graph[i]
      i += 1
      x = self.find(parent, u)
      y = self.find(parent, v)

      if x != y:
        e += 1
        result.append([u, v, w])
        self.union(parent, rank, x, y)

    minCost = 0
    print("edges in MST")
    for u, v, weight in result:
      minCost += weight
      print("%d -- %d == %d" % (u, v, weight))
    print("minimum spanning tree:", minCost)


g = Graph(4)
g.addEdge(0, 1, 10)
g.addEdge(0, 2, 6)
g.addEdge(0, 3, 5)
g.addEdge(1, 3, 15)
g.addEdge(2, 3, 4)
g.kruskal()

```

### Dijkstra's Algorithm

https://bradfieldcs.com/algos/graphs/dijkstras-algorithm/

```python
import heapq


def calculate_distances(graph, starting_vertex):
    distances = {vertex: float('infinity') for vertex in graph}
    distances[starting_vertex] = 0

    pq = [(0, starting_vertex)]
    while len(pq) > 0:
        current_distance, current_vertex = heapq.heappop(pq)

        # Nodes can get added to the priority queue multiple times. We only
        # process a vertex the first time we remove it from the priority queue.
        if current_distance > distances[current_vertex]:
            continue

        for neighbor, weight in graph[current_vertex].items():
            distance = current_distance + weight

            # Only consider this new path if it's better than any path we've
            # already found.
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))

    return distances


example_graph = {
    'U': {'V': 2, 'W': 5, 'X': 1},
    'V': {'U': 2, 'X': 2, 'W': 3},
    'W': {'V': 3, 'U': 5, 'X': 3, 'Y': 1, 'Z': 5},
    'X': {'U': 1, 'V': 2, 'W': 3, 'Y': 1},
    'Y': {'X': 1, 'W': 1, 'Z': 1},
    'Z': {'W': 5, 'Y': 1},
}
print(calculate_distances(example_graph, 'X'))
```

## Dynamic Programming

https://www.geeksforgeeks.org/program-for-nth-fibonacci-number/

```python
# iterative
def fibonacci1(n):
  a = 0
  b = 1
  if n < 0:
    print('invalid input')
  elif n == 0:
    return a
  elif n == 1:
    return b
  else:
    for i in range(2, n+1):
      c = a + b
      a = b
      b = c
    return b
print(fibonacci1(9))

# recursive
def fibonacci2(n):
  if n <= 1:
    return n
  return fibonacci2(n-1) + fibonacci2(n-2)
n = 9; print(n, 'th fibonacci number: ', fibonacci2(n))

# DP
# cut out overlapping subproblems to speed computation
def fibonacci3(n):
  f = [0, 1] # define first two fibonacci numbers
  for i in range(2, n+1):
    f.append(f[i-1] + f[i-2])
  return f[n]
print(fibonacci3(9))

# as you see DP mimicks recursive function but optimised by only repeating what's needed
```
